```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 假设您的数据已经被加载到名为 `data` 的DataFrame中，并且 'abstract_clean' 列包含了文本数据
text_for_wordcloud = ' '.join(data['abstract_clean'].dropna())

# 创建词云对象，设置较高的分辨率
wordcloud = WordCloud(
    width=1600,  # 设置词云的宽度
    height=800,  # 设置词云的高度
    background_color='white',  # 设置背景颜色
    max_words=200,  # 设置最大显示的词数
    max_font_size=150,  # 设置字体最大值
    scale=3,  # 调整词云图的比例，提高清晰度
    random_state=42  # 设置有固定随机种子的随机状态，以便每次生成的词云都是相同的
).generate(text_for_wordcloud)

# 设置图像
plt.figure(figsize=(20, 10))  # 设置图像大小
plt.imshow(wordcloud, interpolation="bilinear")  # 使用双线性插值显示更平滑的图像
plt.axis('off')  # 不显示坐标轴
plt.tight_layout(pad=0)  # 设置布局紧凑

# 保存词云图像到文件，可选
# wordcloud.to_file('high_resolution_wordcloud.png')

# 显示图像
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载数据
data = pd.read_csv('/Users/lianghaosmacbook/Desktop/北京大学博一/数据可视化/可视化大作业/arxiv-ai-analysis-master/processed_data/clean_data.csv')

# Since the data is already filtered, we assume the 'primary_cat' column only contains 'cs:AI', 'cs:LG', 'stat:ML'

# Convert 'created' column to datetime format and extract the year
data['year'] = pd.to_datetime(data['created'], format='%Y/%m/%d').dt.year

# Group by year and 'primary_cat' and count the number of papers
yearly_counts = data.groupby(['year', 'primary_cat']).size().unstack(fill_value=0)

# Now let's plot the data
plt.figure(figsize=(15, 7))
for category in yearly_counts.columns:
    plt.plot(yearly_counts.index, yearly_counts[category], marker='o', label=category)

plt.title('Number of Papers by Year in AI, Machine Learning, and Statistics Categories')
plt.xlabel('Year')
plt.ylabel('Number of Papers')
plt.legend(title='Category')
plt.grid(True)
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/Users/lianghaosmacbook/Desktop/北京大学博一/数据可视化/可视化大作业/arxiv-ai-analysis-master/processed_data/clean_data.csv')

# Convert 'created' column to datetime format and extract the year
data['year'] = pd.to_datetime(data['created'], format='%Y/%m/%d').dt.year

# Group by year and 'primary_cat' and count the number of papers
yearly_counts = data.groupby(['year', 'primary_cat']).size().unstack(fill_value=0)

# Now let's plot the data with enhanced styling
plt.figure(figsize=(15, 7))
line_styles = ['-', '--', '-.']  # Define line styles for different categories
for i, category in enumerate(yearly_counts.columns):
    plt.plot(yearly_counts.index, yearly_counts[category], line_styles[i % len(line_styles)],
             marker='o', linewidth=2.5, label=category)  # Increase line width here

# Beautify the plot
plt.title('Number of Papers by Year in AI, Machine Learning, and Statistics Categories', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Papers', fontsize=14)
plt.legend(title='Category', fontsize=12)
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping

# Optionally set a style
plt.style.use('seaborn-dark-palette')

# Save the figure if needed
# plt.savefig('/Users/lianghaosmacbook/Desktop/北京大学博一/数据可视化/可视化大作业/arxiv-ai-analysis-master/processed_data/yearly_papers_plot.png', dpi=300)

# Show the plot
plt.show()
```

```python
import matplotlib.dates as mdates


# Convert 'created' column to datetime format and extract the year and month
data['year_month'] = pd.to_datetime(data['created']).dt.to_period('M')

# Group by year-month and 'primary_cat' and count the number of papers
monthly_counts = data.groupby(['year_month', 'primary_cat']).size().unstack(fill_value=0)

# Convert the PeriodIndex back to datetime to plot it correctly
monthly_counts.index = monthly_counts.index.to_timestamp()

# Now let's plot the data with enhanced styling
plt.figure(figsize=(15, 7))
line_styles = ['-', '--', '-.']  # Define line styles for different categories
for i, category in enumerate(monthly_counts.columns):
    plt.plot(monthly_counts.index, monthly_counts[category], line_styles[i % len(line_styles)],
             marker='o', linewidth=2.5, label=category)  # Increase line width here

# Beautify the plot
plt.title('Number of Papers by Month in AI, Machine Learning, and Statistics Categories', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Papers', fontsize=14)
plt.legend(title='Category', fontsize=12)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Format x-axis labels as year-month
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=12))  # Show one label per year
plt.xticks(rotation=90)  # Rotate x-axis labels to prevent overlap
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping

# Optionally set a style
plt.style.use('seaborn-dark-palette')

# Show the plot
plt.show()
```

```python

# Convert 'created' column to datetime format and extract the year and month
data['year_month'] = pd.to_datetime(data['created']).dt.to_period('M')

# Group by year-month and 'primary_cat' and count the number of papers
monthly_counts = data.groupby(['year_month', 'primary_cat']).size().unstack(fill_value=0)

# Convert the PeriodIndex back to datetime to plot it correctly
monthly_counts.index = monthly_counts.index.to_timestamp()

# Filter for the last five years
last_five_years = monthly_counts.loc[monthly_counts.index >= (monthly_counts.index.max() - pd.DateOffset(years=5))]

# Now let's plot the data for the last five years with enhanced styling
plt.figure(figsize=(15, 7))
line_styles = ['-', '--', '-.']  # Define line styles for different categories
for i, category in enumerate(last_five_years.columns):
    plt.plot(last_five_years.index, last_five_years[category], line_styles[i % len(line_styles)],
             marker='o', linewidth=2.5, label=category)  # Increase line width here

# Beautify the plot
plt.title('Number of Papers by Month in AI, Machine Learning, and Statistics Categories (Last 5 Years)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Papers', fontsize=14)
plt.legend(title='Category', fontsize=12)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # Format x-axis labels as year-month
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))  # Show a label for every month
plt.xticks(rotation=90)  # Rotate x-axis labels to prevent overlap
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping

# Optionally set a style
plt.style.use('seaborn-dark-palette')

# Show the plot
plt.show()
```
