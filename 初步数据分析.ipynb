# 提取abstract的top 10用词并输出json文件并保存

```python
from sklearn.feature_extraction.text import CountVectorizer
# 数据预处理：从日期中提取年份
data = data.dropna(subset=['abstract_clean', 'created'])

data['year'] = pd.to_datetime(data['created']).dt.year

# 设置 CountVectorizer 来提取关键词
vectorizer = CountVectorizer(max_features=10, stop_words='english')

# 分年份统计关键词
keywords_by_year = {}
for year in data['year'].unique():
    yearly_data = data[data['year'] == year]
    word_matrix = vectorizer.fit_transform(yearly_data['abstract_clean'])
    sum_words = word_matrix.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
    keywords_by_year[year] = sorted(words_freq, key=lambda x: x[1], reverse=True)[:10]

# 数据可视化
for year in keywords_by_year:
    words = [word[0] for word in keywords_by_year[year]]
    freqs = [word[1] for word in keywords_by_year[year]]
    plt.figure(figsize=(10, 5))
    plt.bar(words, freqs)
    plt.title(f"Top 10 Keywords in {year}")
    plt.xticks(rotation=45)
    plt.ylabel('Frequency')
    plt.show()
```

```python
from collections import defaultdict, Counter
import json
from collections import defaultdict, Counter
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 假设 data 是您已经有的 DataFrame
data['year'] = pd.to_datetime(data['created']).dt.year

# 初始化 defaultdict 用于存储每年的词频统计
word_counts_by_year = defaultdict(Counter)

# 对每个年份的摘要进行词频统计
for year in data['year'].unique():
    year = int(year)  # 确保年份是原生的 int 类型

    # 使用 CountVectorizer 提取该年份的词频
    vectorizer = CountVectorizer(stop_words='english')
    abstracts = data[data['year'] == year]['abstract_clean'].dropna()
    word_matrix = vectorizer.fit_transform(abstracts)
    sum_words = word_matrix.sum(axis=0)
    
    # 将词频添加到对应年份的 Counter 中
    for word, idx in vectorizer.vocabulary_.items():
        word_counts_by_year[year][word] += sum_words[0, idx]

# 为每个年份提取前10个最常见的词汇
top_words_by_year = {
    year: counts.most_common(10) for year, counts in word_counts_by_year.items()
}

# 转换为所需的格式，确保所有数据类型都是 JSON 可序列化的
formatted_data = {
    year: [[word, int(count)] for word, count in top_words] 
    for year, top_words in top_words_by_year.items()
}

json_data = json.dumps(formatted_data, indent=4)

# 打印结果
print(json.dumps(formatted_data, indent=4))
file_path = '/Users/lianghaosmacbook/Desktop/北京大学博一/数据可视化/可视化大作业/Visual_final/file_1.json'
with open(file_path, 'w') as file:
    file.write(json_data)

file_path
```
